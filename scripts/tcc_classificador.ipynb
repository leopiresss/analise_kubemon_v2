```python
# Calcular métricas de performance para todos os classificadores
print("📊 AVALIAÇÃO DE PERFORMANCE DOS CLASSIFICADORES")
print("=" * 60)

# Extrair dados de teste do dataset
X_test = dataset_carregado['X_test_scaled']  # Usar dados escalados
y_test = dataset_carregado['y_test']

print(f"📋 Dados de teste:")
print(f"   • X_test shape: {X_test.shape}")
print(f"   • y_test length: {len(y_test)}")
print(f"   • Classes no teste: {np.unique(y_test)}")

# Obter modelos treinados
modelos_sucesso = util_classificadores.get_successful_models()

# Dicionário para armazenar métricas
metricas_resultados = {}

print(f"\n🎯 Calculando métricas para {len(modelos_sucesso)} classificadores...")
print("-" * 60)
```

```python
# Loop para calcular métricas de cada classificador
for nome_modelo, dados in modelos_sucesso.items():
    modelo = dados['model']
    
    print(f"\n🔍 Avaliando: {nome_modelo}")
    print("-" * 40)
    
    try:
        # Fazer predições
        y_pred = modelo.predict(X_test)
        
        # Calcular probabilidades (se disponível)
        try:
            y_prob = modelo.predict_proba(X_test)[:, 1]  # Probabilidade da classe positiva
            tem_prob = True
        except:
            y_prob = None
            tem_prob = False
        
        # Calcular métricas
        accuracy = accuracy_score(y_test, y_pred)
        precision = precision_score(y_test, y_pred, average='binary')
        recall = recall_score(y_test, y_pred, average='binary')
        f1 = f1_score(y_test, y_pred, average='binary')
        
        # AUC-ROC (apenas se probabilidades estão disponíveis)
        if tem_prob:
            auc_roc = roc_auc_score(y_test, y_prob)
        else:
            auc_roc = "N/A (sem probabilidades)"
        
        # Armazenar resultados
        metricas_resultados[nome_modelo] = {
            'accuracy': accuracy,
            'precision': precision,
            'recall': recall,
            'f1_score': f1,
            'auc_roc': auc_roc,
            'y_pred': y_pred,
            'y_prob': y_prob
        }
        
        # Exibir métricas formatadas
        print(f"   ✅ Acurácia:  {accuracy:.4f} ({accuracy*100:.2f}%)")
        print(f"   ✅ Precisão:  {precision:.4f} ({precision*100:.2f}%)")
        print(f"   ✅ Recall:    {recall:.4f} ({recall*100:.2f}%)")
        print(f"   ✅ F1-Score:  {f1:.4f} ({f1*100:.2f}%)")
        if tem_prob:
            print(f"   ✅ AUC-ROC:   {auc_roc:.4f} ({auc_roc*100:.2f}%)")
        else:
            print(f"   ⚠️  AUC-ROC:   {auc_roc}")
            
    except Exception as e:
        print(f"   ❌ Erro ao avaliar {nome_modelo}: {str(e)}")
        metricas_resultados[nome_modelo] = {
            'erro': str(e)
        }

print(f"\n🎉 AVALIAÇÃO CONCLUÍDA!")
print(f"📊 {len(metricas_resultados)} classificadores avaliados")
```

```python
# Criar tabela resumo das métricas
print("📋 RESUMO COMPARATIVO DAS MÉTRICAS")
print("=" * 80)

# Criar DataFrame com os resultados
resultados_df = []

for nome_modelo, metricas in metricas_resultados.items():
    if 'erro' not in metricas:
        linha = {
            'Classificador': nome_modelo,
            'Acurácia': f"{metricas['accuracy']:.4f}",
            'Precisão': f"{metricas['precision']:.4f}",
            'Recall': f"{metricas['recall']:.4f}",
            'F1-Score': f"{metricas['f1_score']:.4f}",
            'AUC-ROC': f"{metricas['auc_roc']:.4f}" if isinstance(metricas['auc_roc'], float) else metricas['auc_roc']
        }
        resultados_df.append(linha)

# Converter para DataFrame
df_metricas = pd.DataFrame(resultados_df)

# Exibir tabela
print(df_metricas.to_string(index=False))

# Encontrar melhor modelo para cada métrica
print(f"\n🏆 MELHORES PERFORMANCES:")
print("-" * 40)

for metrica in ['Acurácia', 'Precisão', 'Recall', 'F1-Score']:
    if metrica in df_metricas.columns:
        # Converter para float para comparação
        valores_numericos = pd.to_numeric(df_metricas[metrica], errors='coerce')
        idx_melhor = valores_numericos.idxmax()
        melhor_modelo = df_metricas.iloc[idx_melhor]['Classificador']
        melhor_valor = df_metricas.iloc[idx_melhor][metrica]
        print(f"   🥇 {metrica}: {melhor_modelo} ({melhor_valor})")

# AUC-ROC separadamente (pode ter valores N/A)
auc_numericos = pd.to_numeric(df_metricas['AUC-ROC'], errors='coerce')
if not auc_numericos.isna().all():
    idx_melhor_auc = auc_numericos.idxmax()
    melhor_modelo_auc = df_metricas.iloc[idx_melhor_auc]['Classificador']
    melhor_valor_auc = df_metricas.iloc[idx_melhor_auc]['AUC-ROC']
    print(f"   🥇 AUC-ROC: {melhor_modelo_auc} ({melhor_valor_auc})")
```

```python
import lib_classificador
from feature_analyzer import FeatureAnalyzer
from dataset_manager import DatasetManager
from lib_classificador import UtilClassificadores

import os
os.environ['OMP_NUM_THREADS'] = '1'
os.environ['OPENBLAS_NUM_THREADS'] = '1'
os.environ['MKL_NUM_THREADS'] = '1'

# Configurar NumPy para usar apenas 1 thread
import numpy as np
np.seterr(all='ignore')

def generate_model_metrics_dataset(methods, titles, X_test, y_test, dataset_name, 
                                   save_to_csv=True, save_dir='../dataset', 
                                   display_results=True, verbose=True):
    """
    Gera dataset completo com métricas de avaliação para múltiplos modelos de ML
    
    Parâmetros:
    - methods: lista de objetos classificadores treinados
    - titles: lista de nomes dos classificadores
    - X_test: dados de teste (features)
    - y_test: rótulos de teste (target)
    - dataset_name: nome do dataset para identificação
    - save_to_csv: se True, salva o dataset em arquivo CSV
    - save_dir: diretório para salvar o arquivo CSV
    - display_results: se True, exibe tabelas e estatísticas detalhadas
    - verbose: se True, exibe progresso detalhado
    
    Retorna:
    - pandas.DataFrame: DataFrame com métricas de todos os modelos
    """
    
    from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, classification_report
    import pandas as pd
    import os
    from datetime import datetime
    import numpy as np
    
    if verbose:
        print("📊 GERANDO DATASET COM MÉTRICAS DE AVALIAÇÃO DOS MODELOS")
        print("=" * 70)
        print(f"🎯 Configuração:")
        print(f"   • Número de modelos: {len(methods)}")
        print(f"   • Dataset: {dataset_name}")
        print(f"   • Shape dados teste: {X_test.shape}")
        print(f"   • Classes únicas: {len(np.unique(y_test))}")
        print(f"   • Salvar CSV: {'Sim' if save_to_csv else 'Não'}")
    
    # Lista para armazenar as métricas
    metrics_data = []
    
    # Iterar por todos os modelos treinados
    for i, (clf, title) in enumerate(zip(methods, titles)):
        if verbose:
            print(f"\n🔍 [{i+1}/{len(methods)}] Avaliando modelo: {title}")
        
        try:
            # Fazer predições no conjunto de teste
            y_pred = clf.predict(X_test)
            
            # Calcular métricas principais
            accuracy = accuracy_score(y_test, y_pred)
            precision = precision_score(y_test, y_pred, average='weighted', zero_division=0)
            recall = recall_score(y_test, y_pred, average='weighted', zero_division=0)
            f1 = f1_score(y_test, y_pred, average='weighted', zero_division=0)

            # Adicionar dados à lista
            metrics_data.append({
                'modelo': title,
                'acuracia': accuracy,
                'precisao': precision,
                'recall': recall,
                'f1_score': f1,
                'dataset': dataset_name,
                'data_avaliacao': datetime.now().strftime('%Y-%m-%d %H:%M:%S'),
                'num_amostras_teste': len(y_test),
                'num_classes': len(np.unique(y_test)),
                'status': 'success'
            })
            
            if verbose:
                print(f"   ✅ {title}: Acurácia={accuracy:.4f}, Precisão={precision:.4f}, Recall={recall:.4f}, F1={f1:.4f}")
                
        except Exception as e:
            if verbose:
                print(f"   ❌ Erro ao avaliar {title}: {e}")
            
            # Adicionar dados com erro
            metrics_data.append({
                'modelo': title,
                'acuracia': None,
                'precisao': None,
                'recall': None,
                'f1_score': None,
                'dataset': dataset_name,
                'data_avaliacao': datetime.now().strftime('%Y-%m-%d %H:%M:%S'),
                'num_amostras_teste': len(y_test),
                'num_classes': len(np.unique(y_test)),
                'status': 'error',
                'erro': str(e)
            })

    # Criar DataFrame com as métricas
    metrics_df = pd.DataFrame(metrics_data)
    
    
    return metrics_df

# Exemplo de uso da função
print("🚀 DEMONSTRAÇÃO DA FUNÇÃO generate_model_metrics_dataset()")
print("=" * 75)

# methods, titles = create_classifiers()


dm_novo = DatasetManager('svm_integracao_teste')
datasets = dm_novo.carregar_dataset('../datasets/svm_integracao_teste.pkl')
dm_novo.imprimir_informacoes()
nome_dataset = 'svm'
classificadores = UtilClassificadores(titles=['KNN','ADA', 'Bag'])
trained_models = classificadores.train_classifiers(datasets['X_train_scaled'], datasets['y_train'])
print(trained_models)
#classificadores.evaluate_classifiers(datasets['X_test_scaled'], datasets['y_test'])
methods = classificadores.methods
print("===================")
print(methods)
X_test_scaled = datasets['X_test_scaled']
y_test = datasets['y_test']
titles = classificadores.titles
metrics_data = []
dataset_name = 'svm'
verbose = True
for i, (clf, title) in enumerate(zip(methods, titles)):
    print(f"\n🔍 [{i+1}/{len(methods)}] Avaliando modelo: {title}")
    

    # Fazer predições no conjunto de teste
    y_pred = clf.predict(X_test_scaled)
    
    # Calcular métricas principais
    accuracy = accuracy_score(y_test, y_pred)
    precision = precision_score(y_test, y_pred, average='weighted', zero_division=0)
    recall = recall_score(y_test, y_pred, average='weighted', zero_division=0)
    f1 = f1_score(y_test, y_pred, average='weighted', zero_division=0)

    # Adicionar dados à lista
    metrics_data.append({
        'modelo': title,
        'acuracia': accuracy,
        'precisao': precision,
        'recall': recall,
        'f1_score': f1,
        'dataset': dataset_name,
        'data_avaliacao': datetime.now().strftime('%Y-%m-%d %H:%M:%S'),
        'num_amostras_teste': len(y_test),
        'num_classes': len(np.unique(y_test)),
        'status': 'success'
    })
    

# Criar DataFrame com as métricas
metrics_df = pd.DataFrame(metrics_data)


#titles = classificadores.titles


#print(titles)

"""
scores = []
for method, name in zip(methods, titles):
    method.fit(datasets['X_train_scaled'], datasets['y_train'])
    scores.append(method.score(datasets['X_test_scaled'], datasets['y_test']))
    print("Classification accuracy {} = {}"
          .format(name, method.score(datasets['X_test_scaled'], datasets['y_test'])))

metrics_df = generate_model_metrics_dataset(
    methods=methods,
    titles=titles,
    X_test=datasets['X_test_scaled'],
    y_test=datasets['y_test'],
    dataset_name=nome_dataset,
    save_to_csv=True,
    save_dir='../dataset',
    display_results=True,
    verbose=True
    )

metrics_df
"""
```

```python
# Solução para o erro KNN - 'NoneType' object has no attribute 'split'
print("🔧 RESOLVENDO ERRO DO KNN")
print("=" * 50)

# O erro 'NoneType' object has no attribute 'split' é comum com KNN no Windows
# devido a problemas com bibliotecas OpenBLAS/MKL. Vamos criar uma solução alternativa.

# Configurar variáveis de ambiente para evitar conflitos
import os
os.environ['OMP_NUM_THREADS'] = '1'
os.environ['MKL_NUM_THREADS'] = '1'
os.environ['OPENBLAS_NUM_THREADS'] = '1'

# Importar bibliotecas necessárias
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import AdaBoostClassifier, BaggingClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
from sklearn.metrics import roc_curve, auc, roc_auc_score
import warnings
warnings.filterwarnings('ignore')

# Função para treinar KNN com tratamento de erro
def treinar_knn_seguro(X_train, y_train, n_neighbors=3):
    """Treina KNN com tratamento de erros específicos"""
    try:
        # Tentar KNN padrão primeiro
        knn = KNeighborsClassifier(n_neighbors=n_neighbors, algorithm='brute')
        knn.fit(X_train, y_train)
        return knn, "KNN_Brute"
    except Exception as e1:
        try:
            # Tentar com algoritmo ball_tree
            knn = KNeighborsClassifier(n_neighbors=n_neighbors, algorithm='ball_tree')
            knn.fit(X_train, y_train)
            return knn, "KNN_BallTree"
        except Exception as e2:
            try:
                # Tentar com algoritmo kd_tree
                knn = KNeighborsClassifier(n_neighbors=n_neighbors, algorithm='kd_tree')
                knn.fit(X_train, y_train)
                return knn, "KNN_KDTree"
            except Exception as e3:
                print(f"❌ Todos os algoritmos KNN falharam:")
                print(f"   • Brute: {str(e1)}")
                print(f"   • Ball Tree: {str(e2)}")
                print(f"   • KD Tree: {str(e3)}")
                return None, "KNN_FAILED"

# Testar o KNN com dados do exemplo
print("🧪 Testando KNN com diferentes algoritmos...")

X_train_test = datasets['X_train_scaled']
y_train_test = datasets['y_train']
X_test_test = datasets['X_test_scaled'] 
y_test_test = datasets['y_test']

knn_model, knn_nome = treinar_knn_seguro(X_train_test, y_train_test)

if knn_model is not None:
    print(f"✅ {knn_nome} treinado com sucesso!")
    
    # Testar predição
    try:
        y_pred_knn = knn_model.predict(X_test_test)
        acc_knn = accuracy_score(y_test_test, y_pred_knn)
        print(f"📊 Acurácia do {knn_nome}: {acc_knn:.4f}")
    except Exception as e:
        print(f"❌ Erro na predição: {str(e)}")
else:
    print("❌ Não foi possível treinar nenhuma variante do KNN")
```

```python
# Retreinar classificadores com correções aplicadas
print("🔄 RETREINANDO CLASSIFICADORES COM CORREÇÕES")
print("=" * 60)

# Criar novos classificadores com configurações mais robustas
def criar_classificadores_robustos():
    """Cria classificadores com configurações que evitam erros comuns"""
    
    classificadores_robustos = []
    nomes_robustos = []
    
    # 1. KNN com algoritmo brute (mais estável)
    try:
        knn_robusto = KNeighborsClassifier(n_neighbors=3, algorithm='brute', n_jobs=1)
        classificadores_robustos.append(knn_robusto)
        nomes_robustos.append('KNN_Robust')
        print("✅ KNN robusto criado")
    except Exception as e:
        print(f"❌ KNN: {str(e)}")
    
    # 2. AdaBoost (geralmente mais estável)
    try:
        ada_robusto = AdaBoostClassifier(
            estimator=DecisionTreeClassifier(max_depth=1),
            n_estimators=50,  # Reduzir número para ser mais rápido
            random_state=42
        )
        classificadores_robustos.append(ada_robusto)
        nomes_robustos.append('ADA_Robust')
        print("✅ AdaBoost robusto criado")
    except Exception as e:
        print(f"❌ AdaBoost: {str(e)}")
    
    # 3. Bagging com DecisionTree (evitar KNN no Bagging)
    try:
        bag_robusto = BaggingClassifier(
            estimator=DecisionTreeClassifier(max_depth=5),
            n_estimators=10,  # Reduzir para ser mais rápido
            random_state=42,
            n_jobs=1  # Evitar paralelização que pode causar problemas
        )
        classificadores_robustos.append(bag_robusto)
        nomes_robustos.append('Bag_Robust')
        print("✅ Bagging robusto criado")
    except Exception as e:
        print(f"❌ Bagging: {str(e)}")
    
    return classificadores_robustos, nomes_robustos

# Criar classificadores robustos
classif_robustos, nomes_robustos = criar_classificadores_robustos()

print(f"\n📊 Classificadores robustos criados: {len(classif_robustos)}")
print(f"📋 Nomes: {nomes_robustos}")

# Treinar classificadores robustos
print(f"\n🏋️ Treinando classificadores robustos...")

classif_treinados = {}
for classif, nome in zip(classif_robustos, nomes_robustos):
    print(f"\n🔍 Treinando {nome}...")
    try:
        classif.fit(X_train_test, y_train_test)
        classif_treinados[nome] = {
            'model': classif,
            'status': 'success'
        }
        print(f"   ✅ {nome}: Treinamento bem-sucedido")
    except Exception as e:
        classif_treinados[nome] = {
            'model': classif,
            'status': 'error',
            'error': str(e)
        }
        print(f"   ❌ {nome}: {str(e)}")

print(f"\n📈 Resumo do treinamento:")
sucessos = [k for k, v in classif_treinados.items() if v['status'] == 'success']
erros = [k for k, v in classif_treinados.items() if v['status'] == 'error']

print(f"✅ Sucessos: {len(sucessos)} - {sucessos}")
print(f"❌ Erros: {len(erros)} - {erros}")
```

```python
# Avaliar classificadores corrigidos e gerar gráfico ROC/AUC
print("📊 AVALIAÇÃO DOS CLASSIFICADORES CORRIGIDOS")
print("=" * 60)

# Avaliar cada classificador treinado com sucesso
metricas_corrigidas = {}

for nome, dados in classif_treinados.items():
    if dados['status'] == 'success':
        modelo = dados['model']
        print(f"\n🔍 Avaliando {nome}...")
        
        try:
            # Fazer predições
            y_pred = modelo.predict(X_test_test)
            
            # Calcular métricas básicas
            acc = accuracy_score(y_test_test, y_pred)
            prec = precision_score(y_test_test, y_pred, average='binary', zero_division=0)
            rec = recall_score(y_test_test, y_pred, average='binary', zero_division=0)
            f1 = f1_score(y_test_test, y_pred, average='binary', zero_division=0)
            
            # Tentar calcular probabilidades para ROC
            try:
                if hasattr(modelo, 'predict_proba'):
                    y_prob = modelo.predict_proba(X_test_test)[:, 1]
                elif hasattr(modelo, 'decision_function'):
                    y_prob = modelo.decision_function(X_test_test)
                else:
                    y_prob = None
                
                if y_prob is not None:
                    auc_score = roc_auc_score(y_test_test, y_prob)
                else:
                    auc_score = None
            except Exception as prob_e:
                print(f"   ⚠️ Erro ao calcular probabilidades: {str(prob_e)}")
                y_prob = None
                auc_score = None
            
            # Armazenar métricas
            metricas_corrigidas[nome] = {
                'accuracy': acc,
                'precision': prec,
                'recall': rec,
                'f1_score': f1,
                'auc_score': auc_score,
                'y_pred': y_pred,
                'y_prob': y_prob,
                'model': modelo
            }
            
            # Exibir métricas
            print(f"   ✅ Acurácia: {acc:.4f}")
            print(f"   ✅ Precisão: {prec:.4f}")
            print(f"   ✅ Recall: {rec:.4f}")
            print(f"   ✅ F1-Score: {f1:.4f}")
            if auc_score is not None:
                print(f"   ✅ AUC-ROC: {auc_score:.4f}")
            else:
                print(f"   ⚠️ AUC-ROC: N/A")
                
        except Exception as eval_e:
            print(f"   ❌ Erro na avaliação: {str(eval_e)}")

# Resumo das métricas
print(f"\n📋 RESUMO DAS MÉTRICAS CORRIGIDAS:")
print("=" * 50)

if metricas_corrigidas:
    for nome, metricas in metricas_corrigidas.items():
        print(f"🔹 {nome}:")
        print(f"   Acurácia: {metricas['accuracy']:.4f}")
        print(f"   Precisão: {metricas['precision']:.4f}")
        print(f"   Recall: {metricas['recall']:.4f}")
        print(f"   F1-Score: {metricas['f1_score']:.4f}")
        if metricas['auc_score'] is not None:
            print(f"   AUC-ROC: {metricas['auc_score']:.4f}")
        print()
    
    # Encontrar melhor modelo por acurácia
    melhor_acc = max(metricas_corrigidas.items(), key=lambda x: x[1]['accuracy'])
    print(f"🏆 Melhor acurácia: {melhor_acc[0]} ({melhor_acc[1]['accuracy']:.4f})")
else:
    print("❌ Nenhum classificador foi avaliado com sucesso")
```
