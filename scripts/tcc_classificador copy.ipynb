```python
# Calcular m√©tricas de performance para todos os classificadores
print("üìä AVALIA√á√ÉO DE PERFORMANCE DOS CLASSIFICADORES")
print("=" * 60)

# Extrair dados de teste do dataset
X_test = dataset_carregado['X_test_scaled']  # Usar dados escalados
y_test = dataset_carregado['y_test']

print(f"üìã Dados de teste:")
print(f"   ‚Ä¢ X_test shape: {X_test.shape}")
print(f"   ‚Ä¢ y_test length: {len(y_test)}")
print(f"   ‚Ä¢ Classes no teste: {np.unique(y_test)}")

# Obter modelos treinados
modelos_sucesso = util_classificadores.get_successful_models()

# Dicion√°rio para armazenar m√©tricas
metricas_resultados = {}

print(f"\nüéØ Calculando m√©tricas para {len(modelos_sucesso)} classificadores...")
print("-" * 60)
```

```python
# Loop para calcular m√©tricas de cada classificador
for nome_modelo, dados in modelos_sucesso.items():
    modelo = dados['model']
    
    print(f"\nüîç Avaliando: {nome_modelo}")
    print("-" * 40)
    
    try:
        # Fazer predi√ß√µes
        y_pred = modelo.predict(X_test)
        
        # Calcular probabilidades (se dispon√≠vel)
        try:
            y_prob = modelo.predict_proba(X_test)[:, 1]  # Probabilidade da classe positiva
            tem_prob = True
        except:
            y_prob = None
            tem_prob = False
        
        # Calcular m√©tricas
        accuracy = accuracy_score(y_test, y_pred)
        precision = precision_score(y_test, y_pred, average='binary')
        recall = recall_score(y_test, y_pred, average='binary')
        f1 = f1_score(y_test, y_pred, average='binary')
        
        # AUC-ROC (apenas se probabilidades est√£o dispon√≠veis)
        if tem_prob:
            auc_roc = roc_auc_score(y_test, y_prob)
        else:
            auc_roc = "N/A (sem probabilidades)"
        
        # Armazenar resultados
        metricas_resultados[nome_modelo] = {
            'accuracy': accuracy,
            'precision': precision,
            'recall': recall,
            'f1_score': f1,
            'auc_roc': auc_roc,
            'y_pred': y_pred,
            'y_prob': y_prob
        }
        
        # Exibir m√©tricas formatadas
        print(f"   ‚úÖ Acur√°cia:  {accuracy:.4f} ({accuracy*100:.2f}%)")
        print(f"   ‚úÖ Precis√£o:  {precision:.4f} ({precision*100:.2f}%)")
        print(f"   ‚úÖ Recall:    {recall:.4f} ({recall*100:.2f}%)")
        print(f"   ‚úÖ F1-Score:  {f1:.4f} ({f1*100:.2f}%)")
        if tem_prob:
            print(f"   ‚úÖ AUC-ROC:   {auc_roc:.4f} ({auc_roc*100:.2f}%)")
        else:
            print(f"   ‚ö†Ô∏è  AUC-ROC:   {auc_roc}")
            
    except Exception as e:
        print(f"   ‚ùå Erro ao avaliar {nome_modelo}: {str(e)}")
        metricas_resultados[nome_modelo] = {
            'erro': str(e)
        }

print(f"\nüéâ AVALIA√á√ÉO CONCLU√çDA!")
print(f"üìä {len(metricas_resultados)} classificadores avaliados")
```

```python
# Criar tabela resumo das m√©tricas
print("üìã RESUMO COMPARATIVO DAS M√âTRICAS")
print("=" * 80)

# Criar DataFrame com os resultados
resultados_df = []

for nome_modelo, metricas in metricas_resultados.items():
    if 'erro' not in metricas:
        linha = {
            'Classificador': nome_modelo,
            'Acur√°cia': f"{metricas['accuracy']:.4f}",
            'Precis√£o': f"{metricas['precision']:.4f}",
            'Recall': f"{metricas['recall']:.4f}",
            'F1-Score': f"{metricas['f1_score']:.4f}",
            'AUC-ROC': f"{metricas['auc_roc']:.4f}" if isinstance(metricas['auc_roc'], float) else metricas['auc_roc']
        }
        resultados_df.append(linha)

# Converter para DataFrame
df_metricas = pd.DataFrame(resultados_df)

# Exibir tabela
print(df_metricas.to_string(index=False))

# Encontrar melhor modelo para cada m√©trica
print(f"\nüèÜ MELHORES PERFORMANCES:")
print("-" * 40)

for metrica in ['Acur√°cia', 'Precis√£o', 'Recall', 'F1-Score']:
    if metrica in df_metricas.columns:
        # Converter para float para compara√ß√£o
        valores_numericos = pd.to_numeric(df_metricas[metrica], errors='coerce')
        idx_melhor = valores_numericos.idxmax()
        melhor_modelo = df_metricas.iloc[idx_melhor]['Classificador']
        melhor_valor = df_metricas.iloc[idx_melhor][metrica]
        print(f"   ü•á {metrica}: {melhor_modelo} ({melhor_valor})")

# AUC-ROC separadamente (pode ter valores N/A)
auc_numericos = pd.to_numeric(df_metricas['AUC-ROC'], errors='coerce')
if not auc_numericos.isna().all():
    idx_melhor_auc = auc_numericos.idxmax()
    melhor_modelo_auc = df_metricas.iloc[idx_melhor_auc]['Classificador']
    melhor_valor_auc = df_metricas.iloc[idx_melhor_auc]['AUC-ROC']
    print(f"   ü•á AUC-ROC: {melhor_modelo_auc} ({melhor_valor_auc})")
```

```python
# Visualiza√ß√£o das m√©tricas
print("üìä VISUALIZA√á√ÉO DAS M√âTRICAS")
print("=" * 40)

# Preparar dados para visualiza√ß√£o
modelos = []
accuracies = []
precisions = []
recalls = []
f1_scores = []
auc_rocs = []

for nome_modelo, metricas in metricas_resultados.items():
    if 'erro' not in metricas:
        modelos.append(nome_modelo)
        accuracies.append(metricas['accuracy'])
        precisions.append(metricas['precision'])
        recalls.append(metricas['recall'])
        f1_scores.append(metricas['f1_score'])
        
        # AUC-ROC apenas se for num√©rico
        if isinstance(metricas['auc_roc'], float):
            auc_rocs.append(metricas['auc_roc'])
        else:
            auc_rocs.append(0)  # Placeholder para modelos sem AUC-ROC

# Criar gr√°fico de barras
fig, axes = plt.subplots(2, 3, figsize=(18, 12))
fig.suptitle('M√©tricas de Performance dos Classificadores', fontsize=16, fontweight='bold')

# Definir m√©tricas e posi√ß√µes
metricas_plot = [
    ('Acur√°cia', accuracies, (0, 0)),
    ('Precis√£o', precisions, (0, 1)),
    ('Recall', recalls, (0, 2)),
    ('F1-Score', f1_scores, (1, 0)),
    ('AUC-ROC', auc_rocs, (1, 1))
]

colors = ['#FF6B6B', '#4ECDC4', '#45B7D1', '#96CEB4', '#FFEAA7']

for i, (nome_metrica, valores, pos) in enumerate(metricas_plot):
    ax = axes[pos[0], pos[1]]
    bars = ax.bar(modelos, valores, color=colors[i], alpha=0.8, edgecolor='black', linewidth=1)
    
    # Adicionar valores nas barras
    for bar, valor in zip(bars, valores):
        height = bar.get_height()
        ax.text(bar.get_x() + bar.get_width()/2., height + 0.01,
                f'{valor:.3f}', ha='center', va='bottom', fontweight='bold')
    
    ax.set_title(nome_metrica, fontsize=14, fontweight='bold')
    ax.set_ylabel('Score', fontsize=12)
    ax.set_ylim(0, 1.1)
    ax.grid(axis='y', alpha=0.3)
    
    # Rotacionar nomes dos modelos se necess√°rio
    ax.tick_params(axis='x', rotation=45)

# Remover subplot extra
axes[1, 2].remove()

# Adicionar tabela resumo no espa√ßo vazio
ax_table = fig.add_subplot(2, 3, 6)
ax_table.axis('off')

# Criar dados da tabela
table_data = []
for i, modelo in enumerate(modelos):
    row = [
        modelo,
        f"{accuracies[i]:.3f}",
        f"{precisions[i]:.3f}",
        f"{recalls[i]:.3f}",
        f"{f1_scores[i]:.3f}",
        f"{auc_rocs[i]:.3f}" if auc_rocs[i] > 0 else "N/A"
    ]
    table_data.append(row)

table = ax_table.table(
    cellText=table_data,
    colLabels=['Modelo', 'Acc.', 'Prec.', 'Rec.', 'F1', 'AUC'],
    cellLoc='center',
    loc='center'
)
table.auto_set_font_size(False)
table.set_fontsize(9)
table.scale(1, 1.5)
ax_table.set_title('Resumo das M√©tricas', fontsize=14, fontweight='bold')

plt.tight_layout()
plt.show()

print("‚úÖ Visualiza√ß√£o criada com sucesso!")
```

```python
# Gerar gr√°fico das Curvas ROC (AUC/ROC)
print("üìà GERANDO CURVAS ROC (AUC/ROC)")
print("=" * 50)

plt.figure(figsize=(12, 10))

# Cores para cada modelo
cores_modelos = {
    'DecisionTree': '#FF6B6B',
    'RF': '#4ECDC4', 
    'NaiveBayes': '#45B7D1',
    'XGBoost': '#96CEB4',
    'KNN': '#FFEAA7'
}

# Plotar curva ROC para cada modelo
modelos_plotados = 0
for nome_modelo, metricas in metricas_resultados.items():
    if 'erro' not in metricas and metricas['y_prob'] is not None:
        # Calcular curva ROC
        fpr, tpr, thresholds = roc_curve(y_test, metricas['y_prob'])
        auc_score = auc(fpr, tpr)
        
        # Plotar curva
        plt.plot(fpr, tpr, 
                linewidth=2, 
                label=f'{nome_modelo} (AUC = {auc_score:.3f})',
                color=cores_modelos.get(nome_modelo, f'C{modelos_plotados}'))
        
        modelos_plotados += 1
        print(f"‚úÖ {nome_modelo}: AUC = {auc_score:.4f}")

# Plotar linha de refer√™ncia (classificador aleat√≥rio)
plt.plot([0, 1], [0, 1], 'k--', linewidth=1, alpha=0.8, label='Classificador Aleat√≥rio (AUC = 0.500)')

# Configura√ß√µes do gr√°fico
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('Taxa de Falsos Positivos (1 - Especificidade)', fontsize=12)
plt.ylabel('Taxa de Verdadeiros Positivos (Sensibilidade)', fontsize=12)
plt.title('Curvas ROC - Compara√ß√£o dos Classificadores', fontsize=14, fontweight='bold')
plt.legend(loc="lower right", fontsize=10)
plt.grid(True, alpha=0.3)

# Adicionar informa√ß√µes extras
plt.text(0.6, 0.2, f'Modelos avaliados: {modelos_plotados}', 
         bbox=dict(boxstyle="round,pad=0.3", facecolor="lightgray", alpha=0.8),
         fontsize=10)

plt.tight_layout()
plt.show()

print(f"\nüéâ Gr√°fico ROC gerado com sucesso!")
print(f"üìä {modelos_plotados} modelos plotados no gr√°fico")
```

```python
# Solu√ß√£o para o erro KNN - 'NoneType' object has no attribute 'split'
print("üîß RESOLVENDO ERRO DO KNN")
print("=" * 50)

# O erro 'NoneType' object has no attribute 'split' √© comum com KNN no Windows
# devido a problemas com bibliotecas OpenBLAS/MKL. Vamos criar uma solu√ß√£o alternativa.

# Configurar vari√°veis de ambiente para evitar conflitos
import os
os.environ['OMP_NUM_THREADS'] = '1'
os.environ['MKL_NUM_THREADS'] = '1'
os.environ['OPENBLAS_NUM_THREADS'] = '1'

# Importar bibliotecas necess√°rias
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import AdaBoostClassifier, BaggingClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
from sklearn.metrics import roc_curve, auc, roc_auc_score
import warnings
warnings.filterwarnings('ignore')

# Fun√ß√£o para treinar KNN com tratamento de erro
def treinar_knn_seguro(X_train, y_train, n_neighbors=3):
    """Treina KNN com tratamento de erros espec√≠ficos"""
    try:
        # Tentar KNN padr√£o primeiro
        knn = KNeighborsClassifier(n_neighbors=n_neighbors, algorithm='brute')
        knn.fit(X_train, y_train)
        return knn, "KNN_Brute"
    except Exception as e1:
        try:
            # Tentar com algoritmo ball_tree
            knn = KNeighborsClassifier(n_neighbors=n_neighbors, algorithm='ball_tree')
            knn.fit(X_train, y_train)
            return knn, "KNN_BallTree"
        except Exception as e2:
            try:
                # Tentar com algoritmo kd_tree
                knn = KNeighborsClassifier(n_neighbors=n_neighbors, algorithm='kd_tree')
                knn.fit(X_train, y_train)
                return knn, "KNN_KDTree"
            except Exception as e3:
                print(f"‚ùå Todos os algoritmos KNN falharam:")
                print(f"   ‚Ä¢ Brute: {str(e1)}")
                print(f"   ‚Ä¢ Ball Tree: {str(e2)}")
                print(f"   ‚Ä¢ KD Tree: {str(e3)}")
                return None, "KNN_FAILED"

# Testar o KNN com dados do exemplo
print("üß™ Testando KNN com diferentes algoritmos...")

X_train_test = datasets['X_train_scaled']
y_train_test = datasets['y_train']
X_test_test = datasets['X_test_scaled'] 
y_test_test = datasets['y_test']

knn_model, knn_nome = treinar_knn_seguro(X_train_test, y_train_test)

if knn_model is not None:
    print(f"‚úÖ {knn_nome} treinado com sucesso!")
    
    # Testar predi√ß√£o
    try:
        y_pred_knn = knn_model.predict(X_test_test)
        acc_knn = accuracy_score(y_test_test, y_pred_knn)
        print(f"üìä Acur√°cia do {knn_nome}: {acc_knn:.4f}")
    except Exception as e:
        print(f"‚ùå Erro na predi√ß√£o: {str(e)}")
else:
    print("‚ùå N√£o foi poss√≠vel treinar nenhuma variante do KNN")
```

```python
# Retreinar classificadores com corre√ß√µes aplicadas
print("üîÑ RETREINANDO CLASSIFICADORES COM CORRE√á√ïES")
print("=" * 60)

# Criar novos classificadores com configura√ß√µes mais robustas
def criar_classificadores_robustos():
    """Cria classificadores com configura√ß√µes que evitam erros comuns"""
    
    classificadores_robustos = []
    nomes_robustos = []
    
    # 1. KNN com algoritmo brute (mais est√°vel)
    try:
        knn_robusto = KNeighborsClassifier(n_neighbors=3, algorithm='brute', n_jobs=1)
        classificadores_robustos.append(knn_robusto)
        nomes_robustos.append('KNN_Robust')
        print("‚úÖ KNN robusto criado")
    except Exception as e:
        print(f"‚ùå KNN: {str(e)}")
    
    # 2. AdaBoost (geralmente mais est√°vel)
    try:
        ada_robusto = AdaBoostClassifier(
            estimator=DecisionTreeClassifier(max_depth=1),
            n_estimators=50,  # Reduzir n√∫mero para ser mais r√°pido
            random_state=42
        )
        classificadores_robustos.append(ada_robusto)
        nomes_robustos.append('ADA_Robust')
        print("‚úÖ AdaBoost robusto criado")
    except Exception as e:
        print(f"‚ùå AdaBoost: {str(e)}")
    
    # 3. Bagging com DecisionTree (evitar KNN no Bagging)
    try:
        bag_robusto = BaggingClassifier(
            estimator=DecisionTreeClassifier(max_depth=5),
            n_estimators=10,  # Reduzir para ser mais r√°pido
            random_state=42,
            n_jobs=1  # Evitar paraleliza√ß√£o que pode causar problemas
        )
        classificadores_robustos.append(bag_robusto)
        nomes_robustos.append('Bag_Robust')
        print("‚úÖ Bagging robusto criado")
    except Exception as e:
        print(f"‚ùå Bagging: {str(e)}")
    
    return classificadores_robustos, nomes_robustos

# Criar classificadores robustos
classif_robustos, nomes_robustos = criar_classificadores_robustos()

print(f"\nüìä Classificadores robustos criados: {len(classif_robustos)}")
print(f"üìã Nomes: {nomes_robustos}")

# Treinar classificadores robustos
print(f"\nüèãÔ∏è Treinando classificadores robustos...")

classif_treinados = {}
for classif, nome in zip(classif_robustos, nomes_robustos):
    print(f"\nüîç Treinando {nome}...")
    try:
        classif.fit(X_train_test, y_train_test)
        classif_treinados[nome] = {
            'model': classif,
            'status': 'success'
        }
        print(f"   ‚úÖ {nome}: Treinamento bem-sucedido")
    except Exception as e:
        classif_treinados[nome] = {
            'model': classif,
            'status': 'error',
            'error': str(e)
        }
        print(f"   ‚ùå {nome}: {str(e)}")

print(f"\nüìà Resumo do treinamento:")
sucessos = [k for k, v in classif_treinados.items() if v['status'] == 'success']
erros = [k for k, v in classif_treinados.items() if v['status'] == 'error']

print(f"‚úÖ Sucessos: {len(sucessos)} - {sucessos}")
print(f"‚ùå Erros: {len(erros)} - {erros}")
```

```python
# Avaliar classificadores corrigidos e gerar gr√°fico ROC/AUC
print("üìä AVALIA√á√ÉO DOS CLASSIFICADORES CORRIGIDOS")
print("=" * 60)

# Avaliar cada classificador treinado com sucesso
metricas_corrigidas = {}

for nome, dados in classif_treinados.items():
    if dados['status'] == 'success':
        modelo = dados['model']
        print(f"\nüîç Avaliando {nome}...")
        
        try:
            # Fazer predi√ß√µes
            y_pred = modelo.predict(X_test_test)
            
            # Calcular m√©tricas b√°sicas
            acc = accuracy_score(y_test_test, y_pred)
            prec = precision_score(y_test_test, y_pred, average='binary', zero_division=0)
            rec = recall_score(y_test_test, y_pred, average='binary', zero_division=0)
            f1 = f1_score(y_test_test, y_pred, average='binary', zero_division=0)
            
            # Tentar calcular probabilidades para ROC
            try:
                if hasattr(modelo, 'predict_proba'):
                    y_prob = modelo.predict_proba(X_test_test)[:, 1]
                elif hasattr(modelo, 'decision_function'):
                    y_prob = modelo.decision_function(X_test_test)
                else:
                    y_prob = None
                
                if y_prob is not None:
                    auc_score = roc_auc_score(y_test_test, y_prob)
                else:
                    auc_score = None
            except Exception as prob_e:
                print(f"   ‚ö†Ô∏è Erro ao calcular probabilidades: {str(prob_e)}")
                y_prob = None
                auc_score = None
            
            # Armazenar m√©tricas
            metricas_corrigidas[nome] = {
                'accuracy': acc,
                'precision': prec,
                'recall': rec,
                'f1_score': f1,
                'auc_score': auc_score,
                'y_pred': y_pred,
                'y_prob': y_prob,
                'model': modelo
            }
            
            # Exibir m√©tricas
            print(f"   ‚úÖ Acur√°cia: {acc:.4f}")
            print(f"   ‚úÖ Precis√£o: {prec:.4f}")
            print(f"   ‚úÖ Recall: {rec:.4f}")
            print(f"   ‚úÖ F1-Score: {f1:.4f}")
            if auc_score is not None:
                print(f"   ‚úÖ AUC-ROC: {auc_score:.4f}")
            else:
                print(f"   ‚ö†Ô∏è AUC-ROC: N/A")
                
        except Exception as eval_e:
            print(f"   ‚ùå Erro na avalia√ß√£o: {str(eval_e)}")

# Resumo das m√©tricas
print(f"\nüìã RESUMO DAS M√âTRICAS CORRIGIDAS:")
print("=" * 50)

if metricas_corrigidas:
    for nome, metricas in metricas_corrigidas.items():
        print(f"üîπ {nome}:")
        print(f"   Acur√°cia: {metricas['accuracy']:.4f}")
        print(f"   Precis√£o: {metricas['precision']:.4f}")
        print(f"   Recall: {metricas['recall']:.4f}")
        print(f"   F1-Score: {metricas['f1_score']:.4f}")
        if metricas['auc_score'] is not None:
            print(f"   AUC-ROC: {metricas['auc_score']:.4f}")
        print()
    
    # Encontrar melhor modelo por acur√°cia
    melhor_acc = max(metricas_corrigidas.items(), key=lambda x: x[1]['accuracy'])
    print(f"üèÜ Melhor acur√°cia: {melhor_acc[0]} ({melhor_acc[1]['accuracy']:.4f})")
else:
    print("‚ùå Nenhum classificador foi avaliado com sucesso")
```
