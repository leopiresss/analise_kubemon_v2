```python
# Modelo Preditivo baseado em KNN para classifica√ß√£o de d√≠gitos manuscritos
# Protocolo experimental HOLDOUT (70/30)

# Carregando as fun√ß√µes necess√°rias
from sklearn.neighbors import KNeighborsClassifier
from sklearn.datasets import load_digits
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
import numpy as np
import pandas as pd

# Carregando a base de dados
X,y=load_digits(return_X_y=True)

# Exemplo de uso de protocolo Holdout

# Organizando a base de dados em treinamento (70%) e teste (30%)  
Xtrain, Xtest, ytrain, ytest=train_test_split(X,y,test_size=0.3, random_state=42, stratify=y)
ytrain.shape

# Declarando o classificador KNN
knn=KNeighborsClassifier(n_neighbors=3)

# Treinando o classificador
knn.fit(Xtrain, ytrain)

# Realizando o test
score=knn.score(Xtest, ytest)

# Calculando a taxa de acerto
print("Acur√°cia (taxa de acerto):", score)

# Resultados por exemplo de teste
ypred=knn.predict(Xtest)
#print(ypred)  

# Matriz de confus√£o (apenas para classificadores)
cm=confusion_matrix(ytest, ypred)
print(cm)
cmp=ConfusionMatrixDisplay(cm, display_labels=np.arange(10))
cmp.plot()
```
